Prophet Model sales forecast

```python
# Install the Prophet library (only needed if not already installed)
!pip install prophet

import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Load your sales data
file_path = "/content/sales_count_by_sku_and_date.csv"
sales_data = pd.read_csv(file_path)

# Convert DATE column to datetime
sales_data['DATE'] = pd.to_datetime(sales_data['DATE'])

# Aggregate sales data by SKU and date
aggregated_sales = sales_data.groupby(['DATE', 'SKU'])['sales_count'].sum().reset_index()

# Pivot data to get SKUs as columns for time series analysis
timeseries_data = aggregated_sales.pivot(index='DATE', columns='SKU', values='sales_count').fillna(0)

# Get the top 5 SKUs based on total sales
top_skus = timeseries_data.sum().sort_values(ascending=False).head(5).index

# Prepare a dictionary to store forecast results
forecast_results = {}
future_periods = 30  # Number of days to forecast

for sku in top_skus:
    # Prepare data for the specific SKU
    sku_data = timeseries_data[sku].reset_index()
    sku_data.columns = ['ds', 'y']  # Prophet requires 'ds' for dates and 'y' for values

    # Initialize the Prophet model
    model = Prophet()
    model.fit(sku_data)

    # Create future dates and forecast
    future = model.make_future_dataframe(periods=future_periods)
    forecast = model.predict(future)

    # Store the forecast result
    forecast_results[sku] = (sku_data, forecast)

    # Plot the forecast
    fig = model.plot(forecast)
    plt.title(f"Sales Forecast for SKU: {sku}")
    plt.xlabel("Date")
    plt.ylabel("Sales Count")
    plt.show()

# Save forecast results for all SKUs
for sku, (_, forecast) in forecast_results.items():
    forecast.to_csv(f"{sku}_forecast.csv", index=False)
    print(f"Forecast saved for SKU: {sku}")
```

```python
import pandas as pd
from openpyxl import Workbook
from openpyxl.styles import PatternFill, Alignment, Font
from openpyxl.utils.dataframe import dataframe_to_rows
import re

def transform_to_detailed_template(input_df):
    # Initialize the output list
    output = []

    # Get unique SKUs and sizes for each SKU
    skus = input_df['SKU'].unique()
    for sku in skus:
        # Filter data for the current SKU
        sku_data = input_df[input_df['SKU'] == sku]
        sizes = sku_data['SIZE'].unique()

        # Sort sizes
        size_order = ['XS', 'S', 'M', 'L', 'XL', 'XXL', '3XL']
        sorted_sizes = sorted([s for s in sizes if s not in size_order], key=lambda x: (isinstance(x, str), x))
        sorted_sizes = size_order + sorted_sizes if any(s in size_order for s in sizes) else sorted_sizes

        # Create rows for each metric
        inventory_row = {'SKU': sku, 'Metric': 'AVAILABLE_INVENTORY_LEVELS'}
        forecast_row = {'SKU': '', 'Metric': 'Forecasted Sales (Next 30 Days)'}
        required_row = {'SKU': '', 'Metric': 'Required'}

        # Fill the size columns and compute totals
        for size in sorted_sizes:
            size_data = sku_data[sku_data['SIZE'] == size]
            inventory = size_data['AVAILABLE_INVENTORY_LEVELS'].values[0] if not size_data.empty else 0
            forecast = size_data['Forecasted Sales (Next 30 Days)'].values[0] if not size_data.empty else 0
            required = inventory - forecast

            inventory_row[size] = inventory
            forecast_row[size] = forecast
            required_row[size] = required if required < 0 else None  # Keep None for non-negative values

        # Compute totals for the rows
        inventory_row['Totals'] = sum(inventory_row.get(size, 0) for size in sorted_sizes)
        forecast_row['Totals'] = sum(forecast_row.get(size, 0) for size in sorted_sizes)
        required_row['Totals'] = sum(value for value in required_row.values() if isinstance(value, (int, float)) and value < 0)

        # Append rows to the output
        output.append(inventory_row)
        output.append(forecast_row)
        output.append(required_row)

    # Convert the output to a DataFrame
    formatted_df = pd.DataFrame(output)

    return formatted_df

def sanitize_sheet_title(title):
    # Remove invalid characters from the sheet title
    return re.sub(r'[\\\/\*\[\]:\?]', '_', str(title))

def save_to_excel_by_category(input_df, file_name):
    # Create an Excel workbook
    wb = Workbook()

    # Get unique categories
    categories = input_df['Category'].unique()

    for category in categories:
        # Filter data for the current category
        category_data = input_df[input_df['Category'] == category]
        formatted_df = transform_to_detailed_template(category_data)

        # Add a new sheet for the category with sanitized title
        sheet_title = sanitize_sheet_title(category)
        ws = wb.create_sheet(title=sheet_title)

        # Merge the first row and add category name
        ws.merge_cells(start_row=1, start_column=1, end_row=1, end_column=len(formatted_df.columns) + 1)
        ws.cell(row=1, column=1, value=category).alignment = Alignment(horizontal="center", vertical="center")
        ws.cell(row=1, column=1).font = Font(bold=True, size=14)

        # Write the DataFrame to the worksheet starting from the second row
        for row in dataframe_to_rows(formatted_df, index=False, header=True):
            ws.append(row)

        # Apply formatting
        for i, row in enumerate(ws.iter_rows(min_row=2, max_row=ws.max_row, max_col=ws.max_column), start=2):
            for cell in row:
                if i == 2:
                    # Header row formatting
                    cell.fill = PatternFill(start_color="C6EFCE", end_color="C6EFCE", fill_type="solid")
                    cell.font = Font(bold=True)
                    cell.alignment = Alignment(horizontal="center", vertical="center")
                elif cell.column == 1:
                    # SKU column formatting
                    cell.font = Font(bold=True)
                elif row[1].value == "Required":  # "Required" row formatting
                    cell.fill = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")
                    if isinstance(cell.value, (int, float)) and cell.value < 0:
                        cell.font = Font(color="FF0000", bold=True)
                    elif cell.value is None:
                        cell.value = ''  # Ensure non-negative values are displayed as empty

    # Remove the default sheet created by openpyxl
    if 'Sheet' in wb.sheetnames:
        del wb['Sheet']

    # Save the workbook
    wb.save(file_name)

# Read input data
input_file = "/content/FORECAST_WITH_INVENTORY 1(in).csv"  # Replace with your actual file path
input_df = pd.read_csv(input_file)

# Save the transformed data to an Excel file with formatting, split by category
output_file = "Formatted_Sales_Forecast_By_Category.xlsx"
save_to_excel_by_category(input_df, output_file)

print(f"Formatted Excel file saved as {output_file}")
```
